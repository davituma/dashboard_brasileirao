```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_br = pd.read_csv('campeonato-brasileiro-estatisticas-full.csv')
df_br.tail()
```

```python
df_br = df_br.fillna('Não informado')
df_br.head()
```

```python
df_br['posse_de_bola'].unique()
```

```python
df_br[df_br['partida_id'] == 5358]
```

```python
df_br = df_br.sort_values(by='partida_id', ascending=True)
df_br.head()
```

```python
colunas_verificacao = [
    'chutes', 'chutes_no_alvo', 'posse_de_bola', 'passes',
    'precisao_passes', 'faltas', 'cartao_amarelo', 'cartao_vermelho',
    'impedimentos', 'escanteios'
]
```

```python
df_br['type'] = df_br[colunas_verificacao].apply(
    lambda row: 0 if all((row == 0) | (row == 'Não informado')) else 1, axis=1
)
```

```python
df_validos = df_br[df_br['type'] == 1]
df_invalidos = df_br[df_br['type'] == 0]
df_invalidos.head(100)
```

```python
df_br['clube'].value_counts()
```

```python
df_media = df_br.groupby('clube').mean(numeric_only=True).reset_index()
num_cols = df_media.select_dtypes(include=[np.number]).columns

df_media[num_cols] = df_media[num_cols].applymap(lambda x: np.floor(x + 0.5))
df_media
```

```python
# dicionário com médias por clube
medias_dict = df_media.set_index('clube').to_dict(orient='index')

# aplica substituição somente onde type == 0
for idx, row in df_br[df_br['type'] == 0].iterrows():
    clube = row['clube']
    if clube in medias_dict:
        for col in df_media.columns:
            if col != 'clube':  # não substituir nome do time
                df_br.at[idx, col] = medias_dict[clube][col]

```
